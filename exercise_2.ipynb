In this exercise we use data on **GDP per capita (current US dollars)** for selected countries.

GDP per capita measures the average economic output per person in a country and is commonly used to compare economic development levels between countries.

Data source: World Bank – GDP per capita (current US$)  
https://data.worldbank.org/indicator/NY.GDP.PCAP.CD

### Importing libraries

In this step we import the Python libraries needed to create the visualization.

- **Pandas** is used to structure and manipulate the dataset.
- **Matplotlib** is used to create the bar chart visualization.

```python
import matplotlib.pyplot as plt
import pandas as pd
```

### Creating the dataset

In this step we create a dataset containing GDP per capita values for selected countries.

The data are stored in a pandas DataFrame, which organizes the information in a tabular structure. 
Each row represents a country and the corresponding GDP per capita value.

```python
data = {
    "Country": ["Germany", "Portugal", "Brazil", "India", "Nigeria"], # List of countries
    "GDP per capita (USD)": [52000, 27000, 9800, 2600, 2200] # Corresponding GDP per capita values for each country
}

df = pd.DataFrame(data) # Create a DataFrame from the provided data

df
```

### Creating the bar chart

In this step we create a bar chart to visualize the data.

Bar charts are commonly used to compare numerical values across different categories. 
In this case, each bar represents a country and the height of the bar corresponds to the GDP per capita value.

This visualization allows us to easily compare the economic differences between the selected countries.

```python
plt.figure(figsize=(8,5))

plt.bar(df["Country"], df["GDP per capita (USD)"]) # Create a bar chart with countries on the x-axis and GDP per capita on the y-axis

plt.title("GDP per Capita in Selected Countries") # Set the title of the plot
plt.xlabel("Country") # Set the label for the x-axis
plt.ylabel("GDP per capita (USD)") # Set the label for the y-axis

plt.show() # Display the plot
```

## 1. Visualization Goal and Audience

The goal of this visualization is to compare GDP per capita between countries with different economic development levels.

The visualization highlights economic differences between countries and allows the viewer to quickly identify which countries have higher or lower GDP per capita.

The audience for this visualization could include students, researchers, policymakers, or anyone interested in economic comparisons between countries.

## 2. Visualization Dimensions (Visualization Wheel)

According to Alberto Cairo's visualization wheel, this visualization can be characterized as follows:

- **Abstraction:** The chart uses abstract graphical elements (bars) rather than real-world objects.
- **Functionality:** The design focuses on clearly presenting numerical information.
- **Familiarity:** Bar charts are widely used and easily understood by most audiences.
- **Lightness:** The visualization is simple and allows quick interpretation.
- **Unidimensionality:** The chart represents one numerical variable (GDP per capita) across categorical variables (countries).
- **Originality:** The visualization uses a conventional chart type rather than an innovative format.

```python
import numpy as np
import matplotlib.pyplot as plt

labels = [ 
    "Abstraction",
    "Functionality",
    "Density",
    "Multidimensionality",
    "Originality",
    "Novelty",
    "Decoration",
    "Lightness",
    "Unidimensionality",
    "Familiarity"
] # List of characteristics for the visualization wheel, which will be used as labels for the axes in the radar chart

# Scores from 1 to 5 based on characteristics of the bar chart
values = [5, 5, 2, 1, 1, 1, 1, 4, 5, 5]

values += values[:1] # Close the loop for the radar chart

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist() # Calculate the angles for each axis in the radar chart
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6,6), subplot_kw=dict(polar=True)) # Create a radar chart with polar coordinates

ax.plot(angles, values) # Plot the values on the radar chart
ax.fill(angles, values, alpha=0.25) # Fill the area under the plot with a color and set the transparency

ax.set_xticks(angles[:-1]) # Set the x-ticks to the angles corresponding to each label, excluding the last one which is a duplicate for closing the loop
ax.set_xticklabels(labels) # Set the x-tick labels to the provided labels for each axis

ax.set_yticks([1,2,3,4,5]) # Set the y-ticks to a range from 1 to 5, which corresponds to the scoring system used for the characteristics
ax.set_title("Visualization Wheel Characterization") # Set the title of the radar chart to indicate that it represents the characterization of the visualization wheel based on the provided characteristics and their corresponding scores

plt.show() # Display the radar chart
```

### Interpretation of the Visualization Wheel

The bar chart scores high in abstraction, functionality, familiarity and unidimensionality.

This is expected because bar charts are simple and widely used visualizations that focus on 
clear communication of quantitative comparisons.

The visualization scores low in originality, novelty and decoration because the chart 
uses a conventional design without artistic or decorative elements.

Density and multidimensionality are also low because the chart represents only one 
quantitative variable across a small number of categories.

## Potential Drawbacks (Cairo and Tufte)

Although the visualization is simple and effective, some potential drawbacks should be considered.

According to Edward Tufte, visualizations should maximize the data-ink ratio, meaning that unnecessary visual elements should be minimized.

Another possible issue is the lie factor, which occurs when graphical elements exaggerate differences between values.

According to Alberto Cairo, visualizations should provide sufficient context. Including information about the data source and year helps avoid misinterpretation.


## Graphical Variables Used

The bar chart uses several graphical variables:

- Position: each bar is positioned along the horizontal axis according to the country.
- Size (length): the height of each bar represents the GDP per capita value.
- Color: color distinguishes the bars visually.
- Shape: all bars have the same rectangular shape.

Position and size are effective graphical variables for representing quantitative comparisons.

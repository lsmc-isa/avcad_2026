### 1. Create a binary variable for presence/absence of 'Salmo trutta fario'

```python
# Create binary variable: 1 for presence, 0 for absence
df['Salmo_trutta_fario_presence'] = df['Salmo trutta fario'].apply(lambda x: 1 if x > 0 else 0)

print("Distribution of Salmo trutta fario presence/absence:")
display(df['Salmo_trutta_fario_presence'].value_counts())
```

### 2. Compare Mean Annual Temperature (temp_ann) between presence and absence groups

```python
from scipy import stats
import numpy as np

# Separate 'temp_ann' for presence and absence groups and drop any NaNs
temp_ann_presence = df[df['Salmo_trutta_fario_presence'] == 1]['temp_ann'].dropna()
temp_ann_absence = df[df['Salmo_trutta_fario_presence'] == 0]['temp_ann'].dropna()

print("\n--- Non-Standardized Data Analysis ---")

# Check conditions for Shapiro-Wilk test (at least 4 data points and not all identical)
can_run_shapiro_presence = len(temp_ann_presence) >= 4 and temp_ann_presence.nunique() >= 2
can_run_shapiro_absence = len(temp_ann_absence) >= 4 and temp_ann_absence.nunique() >= 2

alpha = 0.05

shapiro_pvalue_presence = np.nan
shapiro_pvalue_absence = np.nan

if can_run_shapiro_presence:
    shapiro_presence = stats.shapiro(temp_ann_presence)
    shapiro_pvalue_presence = shapiro_presence.pvalue
    print(f"\nShapiro-Wilk test for temp_ann (Presence group): W={shapiro_presence.statistic:.3f}, p={shapiro_pvalue_presence:.3f}")
else:
    print(f"\nShapiro-Wilk test for temp_ann (Presence group): Not applicable (insufficient data or all values identical).")

if can_run_shapiro_absence:
    shapiro_absence = stats.shapiro(temp_ann_absence)
    shapiro_pvalue_absence = shapiro_absence.pvalue
    print(f"Shapiro-Wilk test for temp_ann (Absence group): W={shapiro_absence.statistic:.3f}, p={shapiro_pvalue_absence:.3f}")
else:
    print(f"Shapiro-Wilk test for temp_ann (Absence group): Not applicable (insufficient data or all values identical).")

# Perform appropriate statistical test
if can_run_shapiro_presence and can_run_shapiro_absence and shapiro_pvalue_presence > alpha and shapiro_pvalue_absence > alpha:
    # Both are normal, perform independent t-test
    test_type = 'Independent t-test'
    statistic, p_value = stats.ttest_ind(temp_ann_presence, temp_ann_absence)
else:
    # At least one is not normal or Shapiro not applicable, perform Mann-Whitney U test
    test_type = 'Mann-Whitney U test'
    # Ensure there's enough data for Mann-Whitney U
    if len(temp_ann_presence) > 0 and len(temp_ann_absence) > 0:
        statistic, p_value = stats.mannwhitneyu(temp_ann_presence, temp_ann_absence)
    else:
        statistic, p_value = np.nan, np.nan
        print("Warning: Insufficient data in one or both groups to perform Mann-Whitney U test.")


print(f"\nStatistical Test: {test_type}")
print(f"Statistic: {statistic:.3f}, P-value: {p_value:.3f}")

p_value_non_standardized = p_value
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize temp_ann
scaler = StandardScaler()
df['temp_ann_standardized'] = scaler.fit_transform(df[['temp_ann']])

# Separate standardized 'temp_ann' for presence and absence groups and drop any NaNs
temp_ann_standardized_presence = df[df['Salmo_trutta_fario_presence'] == 1]['temp_ann_standardized'].dropna()
temp_ann_standardized_absence = df[df['Salmo_trutta_fario_presence'] == 0]['temp_ann_standardized'].dropna()

print("\n--- Standardized Data Analysis ---")

# Check conditions for Shapiro-Wilk test (at least 4 data points and not all identical)
can_run_shapiro_std_presence = len(temp_ann_standardized_presence) >= 4 and temp_ann_standardized_presence.nunique() >= 2
can_run_shapiro_std_absence = len(temp_ann_standardized_absence) >= 4 and temp_ann_standardized_absence.nunique() >= 2

shapiro_pvalue_std_presence = np.nan
shapiro_pvalue_std_absence = np.nan

if can_run_shapiro_std_presence:
    shapiro_std_presence = stats.shapiro(temp_ann_standardized_presence)
    shapiro_pvalue_std_presence = shapiro_std_presence.pvalue
    print(f"\nShapiro-Wilk test for standardized temp_ann (Presence group): W={shapiro_std_presence.statistic:.3f}, p={shapiro_pvalue_std_presence:.3f}")
else:
    print(f"\nShapiro-Wilk test for standardized temp_ann (Presence group): Not applicable (insufficient data or all values identical).")

if can_run_shapiro_std_absence:
    shapiro_std_absence = stats.shapiro(temp_ann_standardized_absence)
    shapiro_pvalue_std_absence = shapiro_std_absence.pvalue
    print(f"Shapiro-Wilk test for standardized temp_ann (Absence group): W={shapiro_std_absence.statistic:.3f}, p={shapiro_pvalue_std_absence:.3f}")
else:
    print(f"Shapiro-Wilk test for standardized temp_ann (Absence group): Not applicable (insufficient data or all values identical).")

# Perform appropriate statistical test with standardized data
if can_run_shapiro_std_presence and can_run_shapiro_std_absence and shapiro_pvalue_std_presence > alpha and shapiro_pvalue_std_absence > alpha:
    test_type_std = 'Independent t-test'
    statistic_std, p_value_std = stats.ttest_ind(temp_ann_standardized_presence, temp_ann_standardized_absence)
else:
    test_type_std = 'Mann-Whitney U test'
    if len(temp_ann_standardized_presence) > 0 and len(temp_ann_standardized_absence) > 0:
        statistic_std, p_value_std = stats.mannwhitneyu(temp_ann_standardized_presence, temp_ann_standardized_absence)
    else:
        statistic_std, p_value_std = np.nan, np.nan
        print("Warning: Insufficient data in one or both groups to perform Mann-Whitney U test.")

print(f"\nStatistical Test: {test_type_std}")
print(f"Statistic: {statistic_std:.3f}, P-value: {p_value_std:.3f}")

p_value_standardized = p_value
```

### 1. Identify the 8 most sampled catchments

```python
# Identify the 8 most sampled catchments
top_8_catchments = df['Catchment_name'].value_counts().nlargest(8).index.tolist()

print("Top 8 most sampled catchments:")
print(top_8_catchments)
```

### 2. Filter dataset to include only those catchments

```python
# Filter the DataFrame to include only the top 8 catchments
df_filtered = df[df['Catchment_name'].isin(top_8_catchments)].copy()
```

### 3. Perform one-way ANOVA (f_oneway)

```python
from scipy.stats import f_oneway
import numpy as np

# Prepare data for ANOVA, dropping any NaNs from each group
anova_groups = [df_filtered['Elevation_mean_catch'][df_filtered['Catchment_name'] == catchment].dropna().to_numpy() for catchment in top_8_catchments]

# Filter out any groups that might be empty after dropping NaNs, or have insufficient data or zero variance for ANOVA
valid_anova_groups = []
for group in anova_groups:
    # Ensure group has at least 2 elements and non-zero variance
    if len(group) > 1 and np.var(group) > 0:
        valid_anova_groups.append(group)

# Check if there are at least two valid groups for ANOVA
if len(valid_anova_groups) < 2:
    f_statistic, p_value_anova = np.nan, np.nan
    print("Warning: Not enough valid groups with sufficient data and variance to perform ANOVA.")
else:
    # Perform one-way ANOVA
    f_statistic, p_value_anova = f_oneway(*valid_anova_groups)

print(f"\nOne-way ANOVA for Elevation_mean_catch among top 8 catchments:")
print(f"F-statistic: {f_statistic:.3f}")
print(f"P-value: {p_value_anova:.3f}")

alpha = 0.05
if not np.isnan(p_value_anova):
    if p_value_anova < alpha:
        print(f"Since the p-value ({p_value_anova:.3f}) is less than alpha ({alpha}), we reject the null hypothesis. There is a significant difference in Elevation_mean_catch among the catchments.")
    else:
        print(f"Since the p-value ({p_value_anova:.3f}) is greater than alpha ({alpha}), we fail to reject the null hypothesis. There is no significant difference in Elevation_mean_catch among the catchments.")
else:
    print("ANOVA could not be performed due to insufficient or invalid data in groups.")
```

### 4. Perform Tukey HSD post-hoc test if ANOVA is significant

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

if p_value_anova < alpha:
    print("\nPerforming Tukey HSD post-hoc test:")
    tukey_result = pairwise_tukeyhsd(endog=df_filtered['Elevation_mean_catch'],
                                     groups=df_filtered['Catchment_name'],
                                     alpha=alpha)
    print(tukey_result)

    # Show which pairs are significantly different
    print("\nPairs of catchments with significantly different Elevation_mean_catch (alpha=0.05):")
    significant_pairs = []
    for i, row in enumerate(tukey_result._results_table.data):
        if i > 0 and row[5] == True: # row[5] is the 'reject' column
            significant_pairs.append(f"{row[0]} vs {row[1]}")

    if significant_pairs:
        for pair in significant_pairs:
            print(f"- {pair}")
    else:
        print("No significant pairs found.")
else:
    print("\nANOVA was not significant, therefore no post-hoc test is needed.")
```

### 6. Perform Kruskal-Wallis test (non-parametric equivalent of ANOVA)

```python
from scipy.stats import kruskal

# Perform Kruskal-Wallis H-test
# We use the same valid_anova_groups prepared previously
if len(valid_anova_groups) >= 2:
    h_statistic, p_value_kruskal = kruskal(*valid_anova_groups)
    print(f"\nKruskal-Wallis H-test for Elevation_mean_catch among top 8 catchments:")
    print(f"H-statistic: {h_statistic:.3f}")
    print(f"P-value: {p_value_kruskal:.3f}")

    if p_value_kruskal < alpha:
        print(f"Since the p-value ({p_value_kruskal:.3f}) is less than alpha ({alpha}), we reject the null hypothesis. There is a significant difference in Elevation_mean_catch among the catchments.")
    else:
        print(f"Since the p-value ({p_value_kruskal:.3f}) is greater than alpha ({alpha}), we fail to reject the null hypothesis. There is no significant difference in Elevation_mean_catch among the catchments.")
else:
    h_statistic, p_value_kruskal = np.nan, np.nan
    print("Kruskal-Wallis test could not be performed due to insufficient or invalid data in groups.")
```

Both tests confirmed a **highly significant difference** in mean catchment elevation among the different sampling basins. This suggests that the selected basins have distinct topographical characteristics.
    *   To further investigate this finding, the **Tukey HSD _post-hoc_ Test** (following ANOVA) and the **Dunn's _post-hoc_ Test** (following Kruskal-Wallis) were performed. Both tests identified **numerous significant pairwise differences** between the basins. For example, basins like 'Ebro' and 'Galiza-Norte' showed significant differences with most other basins. However, a few comparisons were not significant, such as 'Cantabrica' vs 'Catala', 'Douro' vs 'Minho', and 'Galiza-Norte' vs 'Guadia'.

The _post-hoc_ tests provided a detailed view of the specific differences between the groups, solidifying the conclusions that environmental conditions (temperature and elevation) vary significantly and can influence the species' presence and the characteristics of the basins.

```python
# Install scikit-posthocs if not already installed
try:
    import scikit_posthocs as sp
except ImportError:
    %pip install scikit-posthocs
    import scikit_posthocs as sp

if not np.isnan(p_value_kruskal) and p_value_kruskal < alpha:
    print("\nPerforming Dunn's post-hoc test:")

    # Prepare data for Dunn's test: a list of NumPy arrays (one array per group) and corresponding labels
    dunn_data_per_group = []
    dunn_group_labels = []

    for catchment_name in top_8_catchments:
        current_group_series = df_filtered[df_filtered['Catchment_name'] == catchment_name]['Elevation_mean_catch'].dropna()

        # Ensure group has at least 2 elements and non-zero variance (more than one unique value)
        if len(current_group_series) > 1 and current_group_series.nunique() > 1:
            dunn_data_per_group.append(current_group_series.to_numpy())  # Append NumPy array for the group
            dunn_group_labels.append(catchment_name)

    if len(dunn_group_labels) < 2:
        print("Warning: Not enough valid groups with sufficient data and variance for Dunn's post-hoc test. Skipping post-hoc test.")
    else:
        try:
            # Perform Dunn's test using the list of arrays format
            # When 'x' is a list of arrays, 'g' argument should be None
            dunn_results = sp.posthoc_dunn(dunn_data_per_group, p_adjust='bonferroni')

            # Map the numerical indices/columns of the result DataFrame back to catchment names
            dunn_results.index = dunn_group_labels
            dunn_results.columns = dunn_group_labels

            print("\nDunn's Test p-values (Bonferroni-corrected):")
            print(dunn_results)

            # Extract significant pairs
            significant_dunn_pairs = []
            for i in range(len(dunn_results.index)):
                for j in range(i + 1, len(dunn_results.columns)):
                    if dunn_results.iloc[i, j] < alpha:
                        group1 = dunn_results.index[i]
                        group2 = dunn_results.columns[j]
                        significant_dunn_pairs.append(f"{group1} vs {group2}")

            if significant_dunn_pairs:
                print("\nPairs of catchments with significantly different Elevation_mean_catch (Dunn's test, alpha=0.05):")
                for pair in significant_dunn_pairs:
                    print(f"- {pair}")
            else:
                print("No significant pairs found with Dunn's test.")
        except Exception as e:
            print(f"An error occurred during Dunn's test calculation: {e}")
            print("Dunn's test could not be performed.")
else:
    print("\nKruskal-Wallis test was not significant or could not be performed, therefore no post-hoc test is needed.")
```

```python
from scipy.stats import f_oneway
import numpy as np

# Prepare data for ANOVA, dropping any NaNs from each group
anova_groups = [df_filtered['Elevation_mean_catch'][df_filtered['Catchment_name'] == catchment].dropna().to_numpy() for catchment in top_8_catchments]

# Filter out any groups that might be empty after dropping NaNs, or have insufficient data or zero variance for ANOVA
valid_anova_groups = []
for group in anova_groups:
    # Ensure group has at least 2 elements and non-zero variance
    if len(group) > 1 and np.var(group) > 0:
        valid_anova_groups.append(group)

# Check if there are at least two valid groups for ANOVA
if len(valid_anova_groups) < 2:
    f_statistic, p_value_anova = np.nan, np.nan
    print("Warning: Not enough valid groups with sufficient data and variance to perform ANOVA.")
else:
    # Perform one-way ANOVA
    f_statistic, p_value_anova = f_oneway(*valid_anova_groups)

print(f"\nOne-way ANOVA for Elevation_mean_catch among top 8 catchments:")
print(f"F-statistic: {f_statistic:.3f}")
print(f"P-value: {p_value_anova:.3f}")

alpha = 0.05
if not np.isnan(p_value_anova):
    if p_value_anova < alpha:
        print(f"Since the p-value ({p_value_anova:.3f}) is less than alpha ({alpha}), we reject the null hypothesis. There is a significant difference in Elevation_mean_catch among the catchments.")
    else:
        print(f"Since the p-value ({p_value_anova:.3f}) is greater than alpha ({alpha}), we fail to reject the null hypothesis. There is no significant difference in Elevation_mean_catch among the catchments.")
else:
    print("ANOVA could not be performed due to insufficient or invalid data in groups.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd
import numpy as np # Ensure numpy is imported for np.nan, np.var

if not np.isnan(p_value_anova) and p_value_anova < alpha:
    print("\nPerforming Tukey HSD post-hoc test:")

    # Reconstruct the data for Tukey HSD from the groups that were valid for ANOVA
    tukey_data_list = []
    tukey_groups_list = []

    # Iterate through the original anova_groups (pandas Series) and top_8_catchments
    # Use the same filtering logic applied to create valid_anova_groups, plus nunique check
    for i, group_series in enumerate(anova_groups): # anova_groups contains pandas Series after dropna()
        # Ensure group has at least 2 elements and non-zero variance (which implies nunique > 1)
        if len(group_series) > 1 and group_series.nunique() > 1:
            catchment_name = top_8_catchments[i]
            tukey_data_list.extend(group_series.tolist()) # Convert series to list and extend
            tukey_groups_list.extend([catchment_name] * len(group_series))

    if len(tukey_groups_list) == 0 or len(set(tukey_groups_list)) < 2:
        print("Warning: Not enough valid groups or observations with variance for Tukey HSD post-hoc test. Skipping post-hoc test.")
    else:
        try:
            # Create a temporary DataFrame for Tukey HSD
            tukey_df_final = pd.DataFrame({
                'Elevation_mean_catch': tukey_data_list,
                'Catchment_name': tukey_groups_list
            })

            tukey_result = pairwise_tukeyhsd(endog=tukey_df_final['Elevation_mean_catch'],
                                             groups=tukey_df_final['Catchment_name'],
                                             alpha=alpha)
            print(tukey_result)

            print("\nPairs of catchments with significantly different Elevation_mean_catch (alpha=0.05):")
            significant_pairs = []
            for row in tukey_result.summary().data[1:]: # Iterate over the summary table data, skipping the header
                if row[-1] == True: # Check the 'reject' column (last element in each row)
                    significant_pairs.append(f"{row[0]} vs {row[1]}")

            if significant_pairs:
                for pair in significant_pairs:
                    print(f"- {pair}")
            else:
                print("No significant pairs found.")
        except Exception as e:
            print(f"An error occurred during Tukey HSD calculation: {e}")
            print("Tukey HSD could not be performed.")
else:
    print("\nANOVA was not significant or could not be performed, therefore no post-hoc test is needed.")
```
